```python
data = {}
passed = [1,10,11,12]
for file in files:
    cfg = yaml.load(open(file, "r"), Loader=yaml.FullLoader)
    num = file.split("_")[4]
    num = num.split(".")[0]
    num = int(num)
    pas = num in passed
    if True: #not pas:
        data[file] = cfg['clock']
        data[file]['out_dividers'].sort()
        data[file]['passed'] = pas
        data[file]['sysref'] = cfg['clock']['output_clocks']['dac_sysref']['rate']
        for clk in cfg['clock']['output_clocks']:
            data[file][clk] = cfg['clock']['output_clocks'][clk]['rate']
        del data[file]['output_clocks']
        data[file]['sysref_div'] = 7812500.0/data[file]['sysref']
print(cfg['clock'])
```
